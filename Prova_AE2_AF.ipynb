```python
import sympy as sy
import pandas as pd
from IPython.display import Image
```

## Parametros

```python
A, B, C, D, E, I = sy.symbols('A B C D E I')
```

```python
P = A+B+C-D
```

```python
q = 5*A+3*B-4*D
```

```python
h = (B+C-D+2)/5
```

```python
L = (A+C-D+7)/5
```

# Questao 1 _ Portico

```python
Image('Screenshot 2020-12-07 201754.jpg',width="50%")
```

## MetDesl

```python
k11 = 4*E*I/h + 4*E*I/L
k12 = 2*E*I/h
```

```python
beta = q*L**2/2
```

## Resultados

```python
q1D1 = -k11*beta/(k11**2-k12**2)
```

```python
q1D2 = -k12*beta/(k11**2-k12**2)
```

```python
Mc = 2*E*I/L*q1D1
Me = 2*E*I/L*q1D2
```

```python
alpha = 6*E*I/L**2*q1D1 + 6*E*I/L**2*q1D2
```

```python
Rcx = 1/h*(Mc+Me+3*q*L**2/2)+alpha
# Rex = -Rcx
```

```python
Rcy = -6*E*I/L**2*q1D1 + q*L**2/2+alpha/2
Rey = -6*E*I/L**2*q1D2 + q*L**2/2+alpha/2
```

# Questao 2 _ Grelha

```python
Image('Screenshot 2020-12-07 201849.jpg',width="50%")
```

## Resultados

```python
R1 = 2*P+q*L
R2 = -2*P
R3 = P+q*L
```

# Alunos

## Resultados por Aluno

```python
def parms(matricula):
    print("\tP: ",P.subs(matricula).evalf(3))
    print("\tq: ",q.subs(matricula).evalf(3))
    print("\tL: ",L.subs(matricula).evalf(3))
    print("\th: ",h.subs(matricula).evalf(3))
```

```python
def q1(matricula):
    print("Q01.")
    print("\tD1: ",q1D1.subs(matricula).evalf(3))
    print("\tD2: ",q1D2.subs(matricula).evalf(3))
    print("\tMc: ",Mc.subs(matricula).evalf(3))
    print("\tMe: ",Me.subs(matricula).evalf(3))
    print("\tRcx: ",Rcx.subs(matricula).evalf(3))
    print("\tRcy: ",Rcy.subs(matricula).evalf(3))
    print("\tRey: ",Rey.subs(matricula).evalf(3))
```

```python
def q2(matricula):
    print("Q02.")
    print("\tR1: ",R1.subs(matricula).evalf(3))
    print("\tR2: ",R2.subs(matricula).evalf(3))
    print("\tR3: ",R3.subs(matricula).evalf(3))
```

```python
def resps(mat):
    print(mat)
    parms(mat)
    q1(mat)
    q2(mat)
```
