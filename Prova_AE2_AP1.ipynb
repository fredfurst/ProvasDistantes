```python
import sympy as sy
import pandas as pd
from IPython.display import Image
```

# Questoes 1 e 2 _ Treliça

## Vars

```python
A, B, C, x = sy.symbols('A B C x')
```

```python
Image('Annotation 2020-09-10 005056.jpg',width="50%")
```

## Momentos

```python
F = 10*(A+B+2)
```

```python
L = (A/9+B*C/23 +2)
```

```python
h = (2*C+5)/5
```

## Tabela

```python
Li = sy.sqrt(L**2+h**2)
rL = sy.sqrt(1+L**2/h**2)
rh = sy.sqrt(1+h**2/L**2)
```

```python
Rbx = F/2/(L/h+L**4/h**4+1)
Rax = -Rbx
Ray = Rby = F/2
T1 = T4 = (-F/4*rL+Rbx/2*rh)
T2 = T5 = (+F/4*rL+Rbx/2*rh)
T3 = F/2+Rbx*(-h/L)
```

# Questões 3 e 4

```python
Image('Annotation 2020-09-11 011829.jpg',width="50%")
```

```python
F1 = 2*(A+B-2*C)
F2 = 2*(2*A-B-C)
L1 = A/5+B*C/10
L2 = C/5+A*B/10
M1 = C+2
```

```python
Rb = ((F1-F2)*(L1+L2)-M1)/L2/L2*(3*L1+2*L2)+2*(F1+F2)
Ra = F1 -M1/2/L2 +(F1-F2)*L1/2/L2 -Rb/2
Rc = F2 +M1/2/L2 -(F1-F2)*L1/2/L2 -Rb/2
Mesq = -F1*L1 +(Ra-F1)*L2
Mdir = -F2*L1 +(Rc-F2)*L2
```

# Alunos

## Resultados por Aluno

```python
def q12(matricula):
    print("Q01. e Q02.")
    print("\tF: ",F.subs(matricula).evalf(3))
    print("\tL: ",L.subs(matricula).evalf(3))
    print("\th: ",h.subs(matricula).evalf(3))
    print("\tRax: ",Rax.subs(matricula).evalf(3))
    print("\tRay = Rby: ",Ray.subs(matricula).evalf(3))
    print("\tRbx: ",Rbx.subs(matricula).evalf(3))
    print("\tT1 = T4: ",T1.subs(matricula).evalf(3))
    print("\tT2 = T5: ",T2.subs(matricula).evalf(3))
    print("\tT3: ",T3.subs(matricula).evalf(3))
```

```python
def q34(matricula):
    print("Q03. e Q04.")
    print("\tF1: ",F1.subs(matricula).evalf(3))
    print("\tF2: ",F2.subs(matricula).evalf(3))
    print("\tL1: ",L1.subs(matricula).evalf(3))
    print("\tL2: ",L2.subs(matricula).evalf(3))
    print("\tM1: ",M1.subs(matricula).evalf(3))
    print("\tRa: ",Ra.subs(matricula).evalf(3))
    print("\tRb: ",Rb.subs(matricula).evalf(3))
    print("\tRc: ",Rc.subs(matricula).evalf(3))
    print("\tMesq: ",Mesq.subs(matricula).evalf(3))
    print("\tMdir: ",Mdir.subs(matricula).evalf(3))
```

```python
def resps(mat):
    q12(mat)
    q34(mat)
```

```python
matricula = [
    (A,4),
    (B,2),
    (C,6),
]
resps(matricula)
```
