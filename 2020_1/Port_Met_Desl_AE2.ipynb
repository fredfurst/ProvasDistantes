```python
import sympy as sy
import numpy as np
```

```python
E,A,I,Lab,Lbc,D1,D2,D3,M1,P2,P3 = sy.symbols('E A I Lab Lbc D1 D2 D3 M1 P2 P3')
```

```python
Image('Annotation 2020-04-24 164603.jpg')
```

```python
stiffness = sy.Matrix([
    [ 4*E*I/Lab+4*E*I/Lbc, 6*E*I/Lbc**2, 6*E*I/Lab**2 ],
    [ 6*E*I/Lbc**2, E*A/Lab+6*E*I/Lbc**3, 0 ],
    [ 6*E*I/Lab**2, 0, E*A/Lbc+6*E*I/Lab**3 ],
])
```

```python
forces = sy.Matrix([
    [ M1 ], # dois termos
    [ P2 ], # termo referente a 50 kN
    [ P3 ], # termo referente a 20 kN
])
```

```python
P50 = 50*4*2**2/6
P50
```

```python
P20 = 20*4*1**2/5
P20
```

```python
inversa = stiffness.inv()
```

### Measurement Units

For coherence in the treatment of measurement units, here's what we have:

Forces (kN)

Stresses (kPa)

Lengths (m)

### Area, Inertia

```python
b = 0.1 # 10 cm
h = 0.2 # 20 cm
```

Area

```python
b*h
```

Inertia

```python
b*h**3/12
```

### Solving

```python
inversa_num = inversa.subs([(E,210000000),(A,b*h),(I,b*h**3/12),(Lab,4),(Lbc,5)])
```

```python
inversa_num
```

#### Displacements

```python
forcas = forces.subs([(M1,10),(P2,0),(P3,0)])
# M1 = 10 kN.m
disps = inversa_num*forcas
```

```python
disps
```
