## Vars

```python
A, B, C, E, I, x = sy.symbols('A B C E I x')
```

## Geometria

```python
Q1_tgalpha = (A+1)/(B+1)
Q1_hipotenuse = sy.sqrt((A+1)**2+(B+1)**2)
Q1_sinalpha = (A+1)/Q1_hipotenuse
Q1_cosalpha = (B+1)/Q1_hipotenuse
```

## Diagramas

Usando 1, 2 e 3 pra mostrar os três trechos?

### Reações

```python
Q1_Rpx = (C+3)
Q1_Rpy = (C+3)*(A+2)/(B+1)
Q1_Rsy = -(C+3)*(A+2)/(B+1)
```

### DEC

```python
Q1_DEC1 = -(C+3)
Q1_DEC2 = -(C+3)*Q1_sinalpha+(C+3)*(A+2)/(B+1)*Q1_cosalpha
Q1_DEC3 = 0
```

### DEN

```python
Q1_DEN1 = -(C+3)*(A+2)/(B+1)
Q1_DEN2 = -(C+3)*Q1_cosalpha-(C+3)*(A+2)/(B+1)*Q1_sinalpha
Q1_DEN3 = (C+3)*(A+2)/(B+1)
```

### DMF

```python
Q1_DMF1 = -(C+3)
Q1_DMF2 = -(C+3)
Q1_DMF3 = 0
```

### Reações

```python
Q3_Rpx = -(A+4)
Q3_Rpy = B-C
Q3_Mp = -5*(C+1)+3*(B+1)+(A+4)
```

### DEN

Separando em trechos 1, 2, 3 e 4.

```python
Q3_DEN1 = -Q3_Rpy
Q3_DEN2 = -Q3_Rpy
Q3_DEN3 = 0
Q3_DEN4 = 0
```

```python
Q3_DEC1 = -Q3_Rpx
Q3_DEC2 = -Q3_Rpx-(A+4)
Q3_DEC3 = Q3_Rpy
Q3_DEC4 = Q3_Rpy-(B+1)
```

## Respostas

### DEN

```python
Q4_DEN1 = -Q4_Rpy
Q4_DEN2 = -Q4_Rpy
Q4_DEN3 = Q4_Fx
Q4_DEN4 = Q4_Fx
Q4_DEN5 = Q4_Fx
```
