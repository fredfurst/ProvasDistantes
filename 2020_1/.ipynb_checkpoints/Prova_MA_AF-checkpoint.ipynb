```python
import sympy as sy
import pandas as pd
from IPython.display import Image
```

```python
A, B, C = sy.symbols('A B C')
```

# Questão 1

```python
Q1_FRx = B-2*C
Q1_FRy = B+6*A
Q1_FR = sy.sqrt(Q1_FRx**2+Q1_FRy**2)
Q1_MR = -5*A*(B+2)-3*C*(A+4)+(B+A)*(C+1)
```

# Questão 2

```python
Q2_FRx = -B-4*C
Q2_FRy = 5*A
Q2_FRz = -B-A
Q2_FR = sy.sqrt(Q2_FRx**2+Q2_FRy**2+Q2_FRz**2)
Q2_MRx = 0
Q2_MRy = -3*C*(A+2)+(B+A)*(C+1)
Q2_MRz = -5*A*(B+2)
Q2_MR = sy.sqrt(Q2_MRx**2+Q2_MRy**2+Q2_MRz**2)
Q2_alpha = sy.acos(Q2_FRx/Q2_FR)/sy.pi*180
Q2_beta = sy.acos(Q2_FRy/Q2_FR)/sy.pi*180
Q2_gamma = sy.acos(Q2_FRz/Q2_FR)/sy.pi*180
Q2_TRx = Q2_MRx*Q2_FRx/Q2_FR
Q2_TRy = Q2_MRy*Q2_FRy/Q2_FR
Q2_TRz = Q2_MRz*Q2_FRz/Q2_FR
Q2_TR = sy.sqrt(Q2_TRx**2+Q2_TRy**2+Q2_TRz**2)
```

# Questão 3

```python
Q3_Ixx = Q3_Iyy = sy.pi*(A+B+C)**4/4+sy.pi*(A+B+C)**4
```

# Questão 4

```python
a = A+B+4
b = (B+2)/2
t = (C+3)/15
Q4_xCG = (a*t/2+(b**2-t**2)/2)/(a+b-t)
Q4_yCG = (a**2/2+(a-t/2)*(b-t))/(a+b-t)
Q4_Ixx = b*a**3/3-(b-t)*(a-t)**3/3
Q4_Iyy = t*b**3/3+(a-t)*t**3/3
```

# Questão 5

```python
u = A+B+2
v = B+1
w = A+B+C+3
Q5_P = 22e3*7.5*0.1*w
Q5_T = (11*w**3/6-5*u**3/3)/v*7.5
Q5_Rx = (11*w**2/2-5*u**2)*7.5-Q5_T
Q5_Ry = Q5_P
```
