```python
import sympy as sy
import pandas as pd
from IPython.display import Image
```

```python
A, B, C = sy.symbols('A B C')
```

# Questão 1

```python
Q1_FRx = -3*B+2*C
Q1_FRy = 2*A+B-C
Q1_FR = sy.sqrt(Q1_FRx**2+Q1_FRy**2)
Q1_MR = 2*A*A-3*B*C-B*B-A*C
```

# Questão 2

```python
Q2_FRx = B-2*C
Q2_FRy = 2*A
Q2_FRz = C-A
Q2_FR = sy.sqrt(Q2_FRx**2+Q2_FRy**2+Q2_FRz**2)
Q2_MRx = 0
Q2_MRy = 3*C*A+A*A-C*C
Q2_MRz = -2*A*B
Q2_MR = sy.sqrt(Q2_MRx**2+Q2_MRy**2+Q2_MRz**2)
Q2_alpha = sy.acos(Q2_FRx/Q2_FR)/sy.pi*180
Q2_beta = sy.acos(Q2_FRy/Q2_FR)/sy.pi*180
Q2_gamma = sy.acos(Q2_FRz/Q2_FR)/sy.pi*180
Q2_TRx = Q2_MRx*Q2_FRx/Q2_FR
Q2_TRy = Q2_MRy*Q2_FRy/Q2_FR
Q2_TRz = Q2_MRz*Q2_FRz/Q2_FR
Q2_TR = sy.sqrt(Q2_TRx**2+Q2_TRy**2+Q2_TRz**2)
```

# Questão 3

```python
T1 = (B+2)*(C+3)/2
T2 = (A+1)*(C+3)
Q3_F = T1+T2
Q3_d = (T1*(C+3)/3+T2*(C+3)/2)/(T1+T2)
```

# Questão 4

```python
a = A+C+4
b = (B+C+2)/2
t = (A+3)/15
Q4_xCG = ((a-2*t)*t*t/2+2*b*t*b/2)/((a-2*t)*t+2*t*(b-t))
Q4_yCG = a/2
Q4_Ixx = b*a**3/12-(b-t)*(a-2*t)**3/12
d1 = b/2-Q4_xCG
d2 = (b-t)/2-Q4_xCG
Q4_Iyy = a*b**3/12+a*b*d1**2 - ((a-2*t)*(b-t)**3/12+(a-2*t)*(b-t)*d2**2)
```

# Questão 5

```python
u = C+2
v = A+1
w = B+4
Q5_T = (50*u-5*v)/(w+v)*sy.sqrt(2)
Q5_Rx = 70+Q5_T/sy.sqrt(2)
Q5_Ry = Q5_T/sy.sqrt(2)
```
